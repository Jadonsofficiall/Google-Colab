# Build a Regression Machine Learning Model using Mama Tee restaurant dataset.

### Importing the necessary librabries

```python
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.model_selection import train_test_split
from xgboost import XGBRegressor
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
```

# Loading the Dataset
#### Using the dataset, which is loaded to the dataframe

```python
tips_df = pd.read_csv('tips.csv')

print("--- First 5 Rows of Mama Tee Restaurant (Tips) Dataset ---")
print(tips_df.head(), "\n")
```

# Data Preprocessing
#### The objective is to predict 'tip' based on total_bill, sex, smoker, day, time, and size.

```python
X = tips_df.drop('tip', axis=1) # Independent variables
y = tips_df['tip']              # Dependent variable (Target)
```

# XGBoost requires numerical inputs. Hence the converting of the categorical text data (sex, smoker, day, time) into numbers.
#### Will use pandas 'get_dummies' for one-hot encoding.

```python
X_encoded = pd.get_dummies(X, drop_first=True)
```

# Performing Data Segmentation
#### Split the data into training and testing sets

```python
X_train, X_test, y_train, y_test = train_test_split(X_encoded, y, test_size=0.2, random_state=42)
```

# Building and Train the XGBoost Model

```python
xgb_model = XGBRegressor(objective='reg:squarederror', random_state=42)
xgb_model.fit(X_train, y_train)
```

# Making Predictions

```python
y_pred = xgb_model.predict(X_test)
```

# Evaluating the Machine Learning Model

```python
r2 = r2_score(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse) # Root Mean Square Error (RMSE) is the square root of MSE.
mae = mean_absolute_error(y_test, y_pred)

print("--- Model Evaluation Metrics ---")
print(f"R Square: {r2:.4f}")
print(f"Mean Square Error (MSE): {mse:.4f}")
print(f"Root Mean Square Error (RMSE): {rmse:.4f}")
print(f"Mean Absolute Error (MAE): {mae:.4f}\n")
```


# Exporting Predictions of Machine Learning Models
#### Writing predictions to a CSV file using the .to_csv() method on a pandas DataFrame.

```python
output_df = pd.DataFrame({
    'Actual_Tip': y_test,
    'Predicted_Tip': np.round(y_pred, 2)
})

output_df.to_csv('tips_xgboost_predictions.csv', index=False)
print("Predictions successfully exported to 'tips_xgboost_predictions.csv'")
```
